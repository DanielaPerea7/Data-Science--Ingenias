14) ¿Cual es la expectativa de vida mas alta y mas baja? Usa la funcion de pandas y la funcion de NumPy y compara el resultado. ¿Que paises tienen esas expectativas de vida?

```python
# Maxima expectativa de vida usando numpy

## Convertir la columna a un array de NumPy
valores_vida = expvida['life_expectancy'].values

## Calcular el máximo ignorando NaN, sino da NaN
expvida_max_numpy = np.nanmax(valores_vida)

print(f"Máxima expectativa de vida (NumPy): {expvida_max_numpy} años")
```

```python
# Maxima expectativa de vida usando pandas
expvida_max_pandas = expvida['life_expectancy'].max()
print(f"Máxima expectativa de vida (Pandas): {expvida_max_pandas} años")
```

```python
# Minima expectativa de vida usando numpy

## Convertir la columna a un array de NumPy
valores_vida = expvida['life_expectancy'].values

## Calcular el mínimo ignorando NaN, sino da NAN
expvida_min_numpy = np.nanmin(valores_vida)

print(f"Mínima expectativa de vida (NumPy): {expvida_min_numpy} años")
```

```python
# Minima expectativa de vida usando pandas
expvida_min_pandas = expvida['life_expectancy'].min()
print(f"Mínima expectativa de vida (Pandas): {expvida_min_pandas} años")
```

```python
# Muestro los resultados maximos
print(f"La expectativa de vida mas alta es de {expvida_max_pandas} años")
```

```python
# Muestro los resultados minimos
print(f"La expectativa de vida mas baja es de {expvida_min_pandas} años")
```

```python
# Filtro las columnas para mostrar solo las filas que tengan valor
# igual al maximo
expvida_filas_max = expvida[expvida['life_expectancy'] == expvida_max_pandas]
print("Paises con la expectativa de vida mas alta:")
print(expvida_filas_max[['Country', 'life_expectancy']])
```

```python
# Filtro las columnas para mostrar solo las filas que tengan valor
# igual al minimo
expvida_filas_min = expvida[expvida['life_expectancy'] == expvida_min_pandas]
print("Paises con la expectativa de vida mas baja:")
print(expvida_filas_min[['Country', 'life_expectancy']])
```

18) ¿Cuantos paises presentaron algun caso de Sarampion (**Measles**) en el 2015? **_Pista_**: Lee acerca de la función [.count()](http://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.count.html?highlight=count#pandas.DataFrame.count)

```python
# Cuento cuantas filas tienen valores distintos de cero

##Filtrar datos del 2015
casos_2015 = expvida[expvida['Year'] == 2015]
##Casos con sarampion >0
casos_sarampion = casos_2015[casos_2015['measles'] > 0]
##Cantidad de paises con sarampion en 2025
cantidad_paises = casos_sarampion['Country'].nunique()

print(f"En 2015, {cantidad_paises} países presentaron al menos un caso de sarampión.")
```
