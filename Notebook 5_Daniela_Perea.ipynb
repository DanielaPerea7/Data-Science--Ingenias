## 1. Arreglo de ceros de longitud 12


```python
import numpy as np

arreglo = np.zeros(12)
print(arreglo)
```

##2. Arreglo de longitud 10 con ceros en todas sus posiciones y un 10 en la posición número 5

```python
# Arreglo de ceros de longitud 10
arreglo = np.zeros(10)

# Asignar el valor 10 en la posición número 5 (índice 4, porque Python empieza en 0)
arreglo[4] = 10

print(arreglo)
```

## 3. Arreglo que tenga los números del 10 al 49 (inclusive)

```python
arreglo = np.arange(10,50)
print(arreglo)
```

## 4. Arreglo 2d de shape (3, 3) que tenga los números del 0 al 8

```python
# Crear un arreglo de 9 elementos del 0 al 8
arreglo_2d = np.arange(9)

#Convertirlo en una matriz de 3x3
arreglo_2d = arreglo_2d.reshape(3,3)

print(arreglo_2d)
```

## 5. Arreglo de números aleatorios de longitud 100 y obtener su media y varianza

```python
#Crear un arreglo aleatorio de longitud 100
arreglo_aleatorio = np.random.rand (100)

# Imprimir los 100 números aleatorios
print("Números aleatorios:")
print(arreglo_aleatorio)

# Calcular la media
media = np.mean(arreglo_aleatorio)

# Calcular la varianza
varianza = np.var(arreglo_aleatorio)

# Mostrar los resultados
print("Media:", media)
print("Varianza:", varianza)
```

## 6. Media de un arreglo usando np.sum



```python
# Crear un arreglo
arreglo_01 = np.array([10, 20, 30, 40, 50])

# Calcular la media usando np.sum
media = np.sum(arreglo_01) / len(arreglo_01)

# Mostrar el resultado
print("Media:", media)
```

## 7. Calcular la varianza de un arreglo usando np.sum y np.mean

```python
#Crear un arreglo
arreglo_02 = np.array([50, 20, 30, 40, 60])

#Calcular la varianza con np.sum y np.mean
# Paso 1: Calcular la media
media = np.mean(arreglo_02)

# Paso 2: Calcular las diferencias al cuadrado
diferencias_cuadrado = (arreglo_02 - media) ** 2

# Paso 3: Calcular la varianza usando np.sum
varianza = np.sum(diferencias_cuadrado) / len(arreglo_02)

# Resumido sería: varianza = np.mean((arreglo_02 - np.mean(arreglo_02)) ** 2)

#Mostrar el resultado
print("Varianza:", varianza)
```
